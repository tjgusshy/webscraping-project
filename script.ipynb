 Webscraping Projects With  Python

 In this project, this site https://books.toscrape.com/ will be scraped.

 The title,price,category, ups, review, and availability.



```python

# importing packages
from bs4 import BeautifulSoup
import requests
import pandas
```

```python


 #creating an empty list 
    all_books = []

# we will be using the try method to catch errors when webscraping the contents
try:
   # iterating in the pages, note that this will only scrape page 1 to 39.
    for i in range(1,40):
        # this is the url for each page
    
        all_page = f'https://books.toscrape.com/catalogue/page-{i}.html'
        # this is the base url
        url = 'https://books.toscrape.com/catalogue/'

        r = requests.get(all_page)

        soup = BeautifulSoup(r.content, 'html.parser')
 
        heading = soup.find_all('h3')


        for i in heading:
            link = i.find('a')['href']
            url_2 =  requests.get(url +link)).text.strip()
            r = requests.get(url_2)
            soup_2 = BeautifulSoup(r.content, 'html.parser')
            title = soup_2.find('h1').text
            price = soup_2.find('p',class_="price_color").text
            category = soup_2.find('ul',class_="breadcrumb").find_all('a')[2].text
            rating = soup_2.find('div', class_="col-sm-6 product_main").find('p',class_="star-rating")['class'][1]
            availability = soup_2.find('p', class_="instock availability")
            review = soup_2.find('table',class_="table table-striped").text.split()[18]
            upc = table = soup_2.find('table',class_="table table-striped").text.split()[0]
            all_books.append([title,price,category,rating,availability,review, upc])
except:
      print('error')
```

```python
# converting to oandas dataframe
df = pandas.DataFrame(all_books,columns=['title','price','category','rating','availability','review', 'upc'])
```

```python
df.to_csv('all_books2.csv')
```
